**PCA:-**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
data=pd.read_csv("D:\Excel R\Assignments\Assignment 8\wine.csv")
```

```python
data
```

```python
data=data.drop(['Type'],axis=1)
data
```

```python
data.describe()
```

```python
data.isna().sum()
```

```python
data_normal=scale(data)
```

```python
data_normal
```

```python
pca=PCA()
pca_values=pca.fit_transform(data_normal)
pca_values
```

```python
### The amount of variance that each PCA explains:-
var=pca.explained_variance_ratio_
var
```

```python
### Cumulative Variance:-
import numpy as np
var1=np.cumsum(np.round(var,decimals=4)*100)
var1
```

```python
### Variance plot for PCA components obtained:-
plt.plot(var1,color='red')
```

```python
pca_values[:,0:1]
```

```python
x = pca_values[:,0]
y = pca_values[:,1]
z = pca_values[:,2]
```

```python
finalDf = pd.concat([pd.DataFrame(pca_values[:,0:3],columns=['pc1','pc2','pc3'])], axis = 1)
```

```python
finalDf
```

**Hierarchical Clustering:-**

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
### Dendrogram:-
plt.figure(figsize=(10,7))
dendrogram=sch.dendrogram(sch.linkage(finalDf,method='complete'))
```

```python
### Creating Clusters:-
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
hc
```

```python
### Saving the clusters for chart:-
y_hc=hc.fit_predict(finalDf)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
Clusters
```

```python
Clusters.value_counts()
```

**K-Means:-**

```python
from sklearn.cluster import KMeans
```

```python
within_clusters=[]
for i in range(1,10):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(finalDf)
    within_clusters.append(kmeans.inertia_)

plt.plot(range(1,10),within_clusters)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within clusters')
plt.show()
```

```python
model=KMeans(n_clusters=3)
model.fit(finalDf)
a=model.labels_
```

```python
finalDf['clusters_kmeans']=model.labels_
finalDf['clusters_hc']=Clusters
```
