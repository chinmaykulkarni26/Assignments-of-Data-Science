```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
data=pd.read_csv("D:\Excel R\Assignments\Assignment 9\my_movies.csv")
```

```python
data
```

```python
d1=data.drop(['V1','V2','V3','V4','V5'],axis=1)
d1
```

```python
d1.info()
```

```python
d1.isna().sum()
```

# Association Rules(i):-

```python
###Apriori Algorithm:-
frequent_itemsets=apriori(d1,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets,metric='confidence',min_threshold=1)
rules
```

```python
import matplotlib.pyplot as plt
rules['support'].plot(kind='line',figsize=(15,10))
```

```python
rules['lift'].plot(kind='line',figsize=(15,10))
```
