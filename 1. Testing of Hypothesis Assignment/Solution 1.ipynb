```python
import pandas as pd
import scipy.stats as st
```

```python
cutlets=pd.read_csv("D:\Excel R\Assignments\Assignment 3\Cutlets.csv")
```

```python
cutlets
```

```python
cutlets.describe()
```

```python
array1=cutlets['Unit A'].values
array2=cutlets['Unit B'].values
```

```python
array1
```

```python
array2
```

```python
stats,p_value=st.ttest_ind(array1,array2)
```

```python
print('Test Statistic(t)=',stats)
print('p_value=',p_value)
```

```python
if p_value>0.05:
    print('There is no significance difference between the diameter of cutlets produced by 2 units!')
else:
    print('There is a significance difference between the diameter of cutlets!')
```
