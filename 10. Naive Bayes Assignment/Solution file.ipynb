```python
columns=['workclass','education','maritalstatus','occupation','relationship','race','sex','native','Salary']
```

```python
train[columns]=train[columns].astype('category')
test[columns]=test[columns].astype('category')
```

```python
train[columns]=train[columns].apply(LabelEncoder().fit_transform)
test[columns]=test[columns].apply(LabelEncoder().fit_transform)
```

```python
train
```

```python
test
```

```python
x_train=train.iloc[:,0:13]
y_train=train.iloc[:,13]
x_test=test.iloc[:,0:13]
y_test=test.iloc[:,13]
```

# Gaussian Naive Bayes:-

```python
gaussian_model=GaussianNB()
fit_gm=gaussian_model.fit(x_train,y_train)
```

```python
y_pred_gm=fit_gm.predict(x_test)
y_pred_gm
```

```python
print(confusion_matrix(y_test,y_pred_gm))
```

```python
accuracy_gm=np.mean(y_pred_gm==y_test)
print('The accuracy of the model using the Gaussian Naive Bayes is:-',accuracy_gm*100,'%')
```

# Multinominal Naive Bayes:-

```python
multi_model=MultinomialNB()
fit_mm=multi_model.fit(x_train,y_train)
```

```python
y_pred_mm=fit_mm.predict(x_test)
y_pred_mm
```

```python
print(confusion_matrix(y_test,y_pred_mm))
```

```python
accuracy_mm=np.mean(y_pred_mm==y_test)
print('The accuracy of the model using Multinominal Naive Bayes is:-',accuracy_mm*100,'%')
```
